```python
from itertools import product
countries = np.unique(np.union1d(df_2003["Reporting country"],df_2003["Partner country"]))
final_df_2003 = pd.DataFrame(list(product(countries, countries)), columns=['Reporting country', 'Partner country'])
final_df_2003["Value"] = 0
```

```python
import math

df_2003_in = df_2003[df_2003["FLOW"]=="IN"]
df_2003_out = df_2003[df_2003["FLOW"]=="OUT"]

for i in final_df_2003["Reporting country"].unique():
    print(i)
    for j in final_df_2003["Partner country"].unique():
        if i != j:
            value_a = df_2003_in.loc[(df_2003_in["Reporting country"]==j)&(df_2003_in["Partner country"]==i), "Value"]
            value_b = df_2003_out.loc[(df_2003_out["Reporting country"]==i)&(df_2003_out["Partner country"]==j), "Value"]
            value_c = df_2003_in.loc[(df_2003_in["Reporting country"]==i)&(df_2003_in["Partner country"]==j), "Value"]
            value_d = df_2003_out.loc[(df_2003_out["Reporting country"]==j)&(df_2003_out["Partner country"]==i), "Value"]
            
            if value_a.isnull().values == False:
                if math.isnan(value_a.item()) == False:
                    if value_a.size == 0:
                        value_a = pd.Series([0])
            else:
                value_a = pd.Series([0])
                
            if value_b.isnull().values == False:
                if math.isnan(value_b.item()) == False:
                    if value_b.size == 0:
                        value_b = pd.Series([0])
            else:
                value_b = pd.Series([0])
                
            if value_c.isnull().values == False:
                if math.isnan(value_c.item()) == False:
                    if value_c.size == 0:
                        value_c = pd.Series([0])
            else:
                value_c = pd.Series([0])
                
            if value_d.isnull().values == False:
                if math.isnan(value_d.item()) == False:
                    if value_d.size == 0:
                        value_d = pd.Series([0])
            else:
                value_d = pd.Series([0])
            
            if value_a.values > 0:
                in_ji = value_a.values
            else:
                in_ji = 0
                
            if value_b.values > 0:
                out_ij = value_b.values
            else:
                out_ij = 0
                
            if value_c.values < 0:
                in_ij = value_c.values*(-1)
            else:
                in_ij = 0
            
            if value_d.values < 0:
                out_ji = value_d.values*(-1)
            else:
                out_ji = 0
                
            value = in_ji + out_ij + in_ij + out_ji
            final_df_2003.loc[(final_df_2003["Reporting country"]==i)&(final_df_2003["Partner country"]==j),"Value"] = value
```
